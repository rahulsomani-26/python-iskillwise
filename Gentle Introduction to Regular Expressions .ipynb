# Some special characters 
# \d stands for decimal 

```python
import re 
text = 'Hello my student roll code is 123A'
pattern = r'\d' 
result = re.search(pattern,text)
# print(result,result.span(),result.group())
```

```python
result
```

```python
import re 
text = 'Hello my student roll code is 123A'
pattern = r'\d\d' 
result = re.search(pattern,text)
print(result,result.span(),result.group())
```

`\d` will match a single digit 

```python
import re 
text = 'Hello my student roll code is 123A'
pattern = r'\d\d\d' 
result = re.search(pattern,text)
print(result,result.span(),result.group())
```

# A . will match any character 

```python
text = 'What is in the name'
pattern = r'.'
result = re.search(pattern,text)
print(result)
```

```python
text = ' What is in the name'
pattern = r'..'
result = re.search(pattern,text)
print(result)
```

### `\w` matches any part of word character and is equivalent to [A-Za-z0-9]

```python
text = '1God has not ears but he hears'
pattern = r'\w'
result = re.search(pattern,text)
print(result)
```

`\s` will match any white space, tab or a newline 

```python
text = 'God has not ears but he hears'
pattern = r'\s'
result = re.search(pattern,text)
print(result)
```

write a function that accepts a string, and returns True if the string consists of any integral value else False 

```python
def is_integer(text:str)->"True if the string represents an integer else False":
    if re.search(r'[0-9]',text):
        return True
    else:
        return False
```

Basic question : Find an email which has a particular format 

```python
MY_MAIL = 'somani.classes@outlook.com'
text = 'Hey you can mail me at {}'.format(MY_MAIL)
```

```python
text
```

```python
pattern = r'\w+\.\w+@\w+\.\w+'
result = re.search(pattern,text)
print(result.group())
```

use of `{}` 

```python
pattern = r'\w+\.\w+@\w+\.\w{2}'
result = re.search(pattern,text)
print(result.group())
```

```python
import re  
text = 'Hello one'
pattern = re.compile(r'\w+')
print(type(pattern), id(pattern))
try:
    print(pattern.__dict__)
except Exception as e:
    print(' No name space for the pattern')
    print(repr(e))
result = re.search(pattern,text)
print(result.group())
```

```python
# Find the number of spaces in a text 

def countspace(text):
    """returns the number of spaces in the {}"""
    pattern = r'\s.'
    result = re.findall(pattern,text)
#     print(result,type(result))
    val = [i.strip() for i in result]
#     return ' The number of spaces in {} is {}'.format(text,len(result))
    return ' {}'.format(val)
```

```python
countspace('I have a big house')
```
